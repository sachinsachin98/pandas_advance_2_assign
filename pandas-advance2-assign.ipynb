In pandas, both loc and iloc are used to access data within a DataFrame, but they have different methods of indexing and selecting data. Here's the difference between the two:

loc:
The loc indexer is label-based, meaning you use column and index labels to select data. It is used to access data based on the labels of rows and columns. This means you can use row and column labels (index and column names) to extract specific data points or slices from the DataFrame.

iloc:
The iloc indexer is integer-based and is used to access data by integer index positions, regardless of the labels. This means you use the row and column integer positions to select data.

```python
#Q3
reindex = [3,0,1,2]
new_df = df.reindex(reindex)

new_df
```

```python
new_df.loc[2:]
```

```python
new_df.iloc[2:3]
```

Yes a difference was observed in both the outputs.
The .loc function uses the label name to reference the row, rather the .iloc function uses the interger index location to reference the rows.

In pandas, window functions are a powerful tool for performing calculations on a specific "window" of data within a DataFrame or Series. These functions are especially useful for tasks like rolling calculations, moving averages, and other operations that involve analyzing a subset of data points within a defined window.

Window functions in pandas are part of the .rolling() method for Series and DataFrames, and they are used in conjunction with various aggregation functions to compute values based on the data within the window.


In pandas, window functions are a powerful tool for performing calculations on a specific "window" of data within a DataFrame or Series. These functions are especially useful for tasks like rolling calculations, moving averages, and other operations that involve analyzing a subset of data points within a defined window.

Window functions in pandas are part of the .rolling() method for Series and DataFrames, and they are used in conjunction with various aggregation functions to compute values based on the data within the window.

Here's how the basic syntax looks:

python
Copy code
window = df['column'].rolling(window_size)
result = window.<aggregation_function>()
In this syntax, window_size is the size of the rolling window, and <aggregation_function> can be replaced with various aggregation functions like mean, sum, min, max, etc.

Types of window functions include:

Simple Moving Average (mean): Calculates the average of data points within a rolling window.

Exponential Moving Average (mean, with min_periods parameter): Computes a weighted moving average, giving more weight to recent data points. It requires specifying the min_periods parameter, which determines the minimum number of data points required to calculate the average.

Sum (sum): Computes the sum of values within the rolling window.

Minimum (min) and Maximum (max): Calculate the minimum and maximum values within the rolling window, respectively.

Standard Deviation (std): Computes the standard deviation of data points within the rolling window.

Variance (var): Computes the variance of values within the rolling window.

Apply Custom Functions (apply): You can use the .apply() method with a custom function to perform calculations within the rolling window.

```python
#Q8
date_1 = input("Enter the first date (YYYY-MM-DD): ")
date_2 = input("Enter the second date (YYYY-MM-DD): ")

date1 = pd.to_datetime(date_1)
date2 = pd.to_datetime(date_2)

time1 = pd.Timedelta(days=3, hours=2, minutes=15)
time2 = pd.Timedelta(days=1, hours=15, minutes=55)

date_time_1 = date1 + time1
date_time_2 = date2 + time2

date_time_diff = date_time_2 - date_time_1

days = date_time_diff.days
hours = date_time_diff.seconds // 3600
minutes = (date_time_diff.seconds // 60) % 60

print(f"The difference between {date1.date()} and {date2.date()} is {days} days, {hours} hours, and {minutes} minutes.")
```

```python
#Q10
import matplotlib.pyplot as plt

filepath = input("Enter file path: ")

df = pd.read_csv(filepath)

sales_by_category = pd.pivot_table(df, index='Date', columns='Product Category', values='Sales', aggfunc='sum')

sales_by_category.plot(kind='bar', stacked=True)

plt.title('Sales by Product Category Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.show()
```
